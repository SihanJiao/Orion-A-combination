# 1 D example for a point source

```python
from astropy.convolution import convolve, Gaussian1DKernel

def gauss_1d(
    data_1d,
    fwhm,
    pixel,
    ):

    FWHM_TO_SIGMA = 1./np.sqrt(8*np.log(2))
    kernel_size=fwhm*FWHM_TO_SIGMA
    pixel_n = kernel_size/pixel
    gauss_1D_kernel =Gaussian1DKernel(pixel_n)
    smoothed_data_gauss = convolve(data_1d, gauss_1D_kernel,normalize_kernel=True)
    
    return smoothed_data_gauss
```

Simulate a 1D point source data and smooth it to high (grey line) and low resolution (red line). For high resolution data, I take out the large scale struction and then get the fake interferometer data (lime line).

```python
plt.figure(figsize=(10, 8))
x = np.linspace(0, 1000, 1001)
y = np.zeros(1001)
y[500] = 1.

# plot high resolution and low resolution data
data_high = gauss_1d(y,10,4.)
data_low = gauss_1d(y,40,4.)

plt.plot(x,data_high,color = (0, 0, 1.0, 0.2), linewidth = 6.,label='high resolution')
plt.plot(x,data_low,color = 'red',linewidth = 3.,label='low resolution')

# plot high resolution data without large scale struction
y = np.linspace(0, 1000, 1001)
rr = ((y-500)**2)**0.5
rr = rr.flatten()
step = (rr>=15) 
data_high_int = np.fft.ifft(np.fft.fft(data_high) * (np.fft.fftshift(step)))
plt.plot(x,data_high_int,color = 'lime',linewidth = 3.,label='high resolution (int)')
plt.xlim(450,550)
legend = plt.legend(loc='upper right', shadow=True, fontsize='large',frameon=False)
plt.xlabel('pixel',fontsize='large')
plt.ylabel('Amplitude',fontsize='large')
```

What we want to do is to combine the low resolution (red line) and interferometer data (lime line) to get high resolution data (grey line).

Step + Gaussian weighting fuction

```python
w_low = (np.fft.fftshift(1-step)*np.fft.fft(data_high)/np.fft.fft(data_low))
fftsum = np.fft.fft(data_low) * w_low + np.fft.fft(data_high_int) * (np.fft.fftshift(step))
plt.figure(figsize=(10, 8))
plt.plot(x,np.fft.fft(data_high_int),color = 'lime',label='high resolution (int)')
plt.plot(x,np.fft.fft(data_low),color = 'red',label='low resolution')
plt.plot(x,np.fft.fft(data_high),color = (0, 0, 1.0, 0.2),label='high resolution')
plt.plot(x,(np.fft.fftshift(1-step)*np.fft.fft(data_high)/np.fft.fft(data_low)),color = 'orange',linewidth = 4.,label='weighting fuction (low)')
plt.plot(x,(np.fft.fftshift(step)),linewidth = 4.,label='weighting fuction (high)',linestyle='dashed')
plt.xlim(0,500)
legend = plt.legend(loc='lower right', shadow=True, fontsize='large',frameon=False)
plt.xlabel('Spatial Frequency',fontsize='large')
plt.ylabel('Amplitude',fontsize='large')
```

```python
combine = np.fft.ifft(fftsum)


fig, axs = plt.subplots(2, 1, sharex=True,figsize=(10, 8))
# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

# Plot each graph, and manually set the y tick values
axs[0].plot(x,data_high,color = (0, 0, 1.0, 0.2), linewidth = 6.,label='high resolution')
axs[0].plot(x,data_low,color = 'red',linewidth = 3.,label='low resolution')
axs[0].plot(x,data_high_int,color = 'lime',linewidth = 3.,label='high resolution (int)')
axs[0].plot(x, combine,color = 'black',label='combined data')
axs[0].set_xlim(450, 550)
legend = axs[0].legend(loc='upper right', shadow=True, fontsize='large',frameon=False)
sub = data_high - combine

axs[1].plot(x, sub)

axs[1].set_xlabel('pixel',fontsize='large')
axs[1].set_ylabel('Amplitude',fontsize='large')
axs[0].set_ylabel('Amplitude',fontsize='large')

```

Feather weighting fuction

```python
#fftsum_feather = np.fft.fft(data_low) * (*np.fft.fft(data_high)/np.fft.fft(data_low)) + np.fft.fft(data_high_int) * (np.fft.fftshift(step))

kfft = np.abs(np.fft.fft(data_low))
kfft /= kfft.max()
plt.figure(figsize=(10, 8))
plt.plot(x,np.fft.fft(data_high_int),color = 'lime',label='high resolution (int)')
plt.plot(x,np.fft.fft(data_low),color = 'red',label='low resolution')
plt.plot(x,np.fft.fft(data_high),color = (0, 0, 1.0, 0.2),label='high resolution')
plt.plot(x,kfft,color = 'orange',linewidth = 4.,label='weighting fuction (low)')
plt.plot(x,(1-kfft),linewidth = 4.,label='weighting fuction (low)',linestyle='dashed')
plt.xlim(0,500)
legend = plt.legend(loc='lower right', shadow=True, fontsize='large',frameon=False)
plt.xlabel('Spatial Frequency',fontsize='large')
plt.ylabel('Amplitude',fontsize='large')
```

```python
fftsum_feather = np.fft.fft(data_low) * kfft + np.fft.fft(data_high_int) * (1-kfft)

combine_feather = np.fft.ifft(fftsum_feather)


fig, axs = plt.subplots(2, 1, sharex=True,figsize=(10, 8))
# Remove horizontal space between axes
fig.subplots_adjust(hspace=0)

# Plot each graph, and manually set the y tick values
axs[0].plot(x,data_high,color = (0, 0, 1.0, 0.2), linewidth = 6.,label='high resolution')
axs[0].plot(x,data_low,color = 'red',linewidth = 3.,label='low resolution')
axs[0].plot(x,data_high_int,color = 'lime',linewidth = 3.,label='high resolution (int)')
axs[0].plot(x, combine_feather,color = 'black',label='combined data (feather)')
axs[0].set_xlim(450, 550)
legend = axs[0].legend(loc='upper right', shadow=True, fontsize='large',frameon=False)

sub_feather = data_high - combine_feather

axs[1].plot(x, sub_feather)

axs[1].set_xlabel('pixel',fontsize='large')
axs[1].set_ylabel('Amplitude',fontsize='large')
axs[0].set_ylabel('Amplitude',fontsize='large')

```

# 2 D example for a point source

```python
from astropy.convolution import convolve, Gaussian2DKernel

def gauss_2d(
    data_2d,
    fwhm,
    pixel,
    ):

    FWHM_TO_SIGMA = 1./np.sqrt(8*np.log(2))
    kernel_size=fwhm*FWHM_TO_SIGMA
    pixel_n = kernel_size/pixel
    gauss_2D_kernel =Gaussian2DKernel(pixel_n)
    smoothed_data_gauss = convolve(data_2d, gauss_2D_kernel,normalize_kernel=True)
    
    return smoothed_data_gauss
```

Simulate a 2D point source data and smooth it to high (grey line) and low resolution (red line). For high resolution data, I take out the large scale struction and then get the fake interferometer data (lime line).

```python
x = np.zeros((401,401))
x[200][200] = 1.

# plot high resolution and low resolution data
data_high = gauss_2d(x,10,1.)
data_low = gauss_2d(x,40,1.)

ygrid, xgrid = np.indices(x.shape, dtype='float')
rr = ((xgrid - x.shape[1]/2)**2+(ygrid - x.shape[0]/2)**2)**0.5
```

```python
ring = (rr >= 20)

data_high_int = np.fft.ifft2(np.fft.fft2(data_high) * (np.fft.fftshift(ring)))
```

```python
fig, axs = plt.subplots(1, 3, sharex=True,figsize=(18, 5))

axs[0].imshow(data_low)

axs[1].imshow(data_high)

axs[2].imshow(data_high_int.real)
```

```python
fftsum = np.fft.fft2(data_low) * (np.fft.fftshift(1-ring)*np.fft.fft2(data_high)/np.fft.fft2(data_low)) + np.fft.fft2(data_high_int) * (np.fft.fftshift(ring))
combine = (np.fft.ifft2(fftsum))
```

```python
kfft = np.abs(np.fft.fft2(np.fft.fftshift(data_low)))
              
fftsum_feather = np.fft.fft2(data_low) * kfft + np.fft.fft2(data_high_int) * (1-kfft)
combine_feather = (np.fft.ifft2(fftsum_feather))
```

```python
fig, axs = plt.subplots(1, 3, sharex=True,figsize=(18, 5))

axs[0].imshow(data_high)

axs[1].imshow(combine.real)

axs[2].imshow(combine_feather.real)

```

```python
fig, axs = plt.subplots(1, 3, sharex=True,figsize=(18, 5))

axs[0].imshow(data_high - data_high_int.real)

axs[1].imshow(data_high - combine.real)

axs[2].imshow(data_high - combine_feather.real)
```
