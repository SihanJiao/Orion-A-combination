```python
import matplotlib
from astropy import units as u
import aplpy
import numpy as np
import astropy.io.fits as fits
from astropy.io import ascii
from astropy.table import Column, Table
%matplotlib inline
```

```python
###########defined the coordinate of reference point
ra_ref = 360*(5.+(35.+33.32839/60.)/60.)/24.
dec_ref = -5.-(6.+15.193627/60.)/60.
ra00 = ra_ref+(-50.)/3600.
dec00 = dec_ref+95./3600.


ra = np.zeros(504)
ra_0 = ra00
r = 0.005 #######radius of observing point
for i in range(0,504):
    z = int(i/18)
    if (z % 2 ==0):
        ra_0 = ra_0 - 2 * r
        ra[i]=ra_0
    else:
        ra_0 = ra_0 + 2 * r
        ra[i]=ra_0
        
    
dec=[]
for i in range(0,504):
    dec.append(float(dec00 + int(i / 18) * (np.sqrt(3) / 2) * r * 2))
```

```python
f = aplpy.FITSFigure('/Users/shjiao/documents/orion/sedlarge/plot/proposal/orion_sma.fits', figsize=(10,10))
f.show_grayscale()

f.add_colorbar()
f.colorbar.show()

f.recenter(83.835,-5.005669,width=0.15,height=0.2)

f.colorbar.set_location('top')
f.colorbar.set_axis_label_text(r'Flux (Jy beam$^{-1}$)')
f.show_colorscale(vmax=0.25,cmap='gist_yarg',stretch='arcsinh')
f.colorbar.set_width(0.15)
f.colorbar.set_ticks([0.05,0.1,0.2])
f.add_beam(major=14.0*u.arcsecond,minor=14.0*u.arcsecond,angle=0.0)
f.add_label(0.17, 0.08, 'Beam: 14.0\"', relative=True, size='large', layer='title',color='black')
f.show_contour('/Users/shjiao/documents/orion/sedlarge/plot/proposal/orion_sma.fits', levels=[0.03], colors='orange')


#####defined your target region
[fitshdr] = fits.open('/Users/shjiao/documents/orion/sedlarge/plot/proposal/orion_sma.fits')
head=fitshdr.header
data=fitshdr.data
xrefpix = head['crpix1']
xrefval = head['crval1']
xwid = head['cdelt1']
yrefpix = head['crpix2']
yrefval = head['crval2']
ywid = head['cdelt2']

xpix = np.ones_like(ra)
ypix = np.ones_like(dec)
for i in range(0,504):
    xpix[i] = (ra[i] - xrefval)/xwid + xrefpix -1
    ypix[i] = (dec[i] - yrefval)/ywid + yrefpix -1

ra_sec = []
dec_sec = []


#######select the observing region based on flux (if need)
for i in range(0,xpix.shape[0]):
    if data[int(ypix[i])][int(xpix[i])] > 0.022 :
        ra_sec.append(float(ra[i]))
        dec_sec.append(float(dec[i]))
        f.show_circles(float(ra[i]),float(dec[i]), 0.005,edgecolors='black',linewidths=0.5)

#######save the csv file
table=Table([ra_sec,dec_sec])
table.write('/Users/shjiao/desktop/za/sma_pointing/omc3_loop1.csv',format='csv',overwrite=True)



f.add_scalebar(0.08333/2.*6./20,"0.1 pc",color='black', corner='top right')

f.tick_labels.set_font(size='large')
f.axis_labels.set_font(size='large')
```
